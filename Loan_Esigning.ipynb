### Splitting the dataset

```python
df.columns
```

```python
x = df.drop(columns = ['entry_id','e_signed'])
```

```python
y = df['e_signed']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
x_test.shape
```

### Feature Scaling

```python
sc = StandardScaler()
```

```python
scaled_x_train = sc.fit_transform(x_train)
scaled_x_test = sc.transform(x_test)
```

```python
scaled_x_train
```

```python
scaled_x_test
```

## Section 2: Model Building

### 1) Logistic Regression

```python
classifier_lr = LogisticRegression(random_state=0)
classifier_lr.fit(scaled_x_train, y_train)
```

```python
y_pred = classifier_lr.predict(scaled_x_test)
```

```python
acc = accuracy_score(y_pred,y_test)
acc
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#Confusion matrix
plt.figure(figsize = (5,5))
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True, cmap = 'Set2', fmt= "d",
            linewidth = 3, cbar = False, xticklabels = ['Yes', 'No'],
            yticklabels = ['Yes', 'No'], square = True
            )
plt.xlabel("True Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.ylabel("Predicted Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.title("CONFUSION MATRIX - LOGISTIC REGRESSION", fontsize = 14, color = 'm')
plt.show()
```

### 2) Support Vector Machines SVM

```python
classifier_svc = SVC(random_state=0)
classifier_svc.fit(scaled_x_train, y_train)
```

```python
y_pred = classifier_svc.predict(scaled_x_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#Confusion matrix
plt.figure(figsize = (5,5))
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True, cmap = 'Set2', fmt= "d",
            linewidth = 3, cbar = False, xticklabels = ['Yes', 'No'],
            yticklabels = ['Yes', 'No'], square = True
            )
plt.xlabel("True Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.ylabel("Predicted Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.title("CONFUSION MATRIX - SVM", fontsize = 14, color = 'm')
plt.show()
```

### 3) Random Forest Classifier

```python
classifier_rf = RandomForestClassifier(random_state=0)
classifier_rf.fit(scaled_x_train, y_train)
```

```python
y_pred = classifier_rf.predict(scaled_x_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#Confusion matrix
plt.figure(figsize = (5,5))
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True, cmap = 'Set2', fmt= "d",
            linewidth = 3, cbar = False, xticklabels = ['Yes', 'No'],
            yticklabels = ['Yes', 'No'], square = True
            )
plt.xlabel("True Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.ylabel("Predicted Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.title("CONFUSION MATRIX - RANDOM FOREST", fontsize = 14, color = 'm')
plt.show()
```

### 4)XGBoost Classifier

```python
classifier_xgb = XGBClassifier(random_state=0)
classifier_xgb.fit(scaled_x_train, y_train)
```

```python
y_pred = classifier_xgb.predict(scaled_x_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#Confusion matrix
plt.figure(figsize = (5,5))
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True, cmap = 'Set2', fmt= "d",
            linewidth = 3, cbar = False, xticklabels = ['Yes', 'No'],
            yticklabels = ['Yes', 'No'], square = True
            )
plt.xlabel("True Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.ylabel("Predicted Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.title("CONFUSION MATRIX - XGBOOST", fontsize = 14, color = 'm')
plt.show()
```

## Section 3: Randomized Search to find the best parameters (XGBoost Classifier)

```python
parameters = { 'learning_rate' : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30],
              'max_depth' : [3, 4, 5, 6, 8, 10, 12, 15],
              'min_child_weight' : [1, 3, 5, 7],
              'gamma' : [0.00, 0.10, 0.20, 0.30, 0.40],
              'colsample_bytree' : [0.3, 0.4, 0.5, 0.7],
              'n_estimators' : [100, 200, 500],
              'subsample' : [0.5, 0.7, 1.0]
}
```

```python
parameters
```

```python
random_search = RandomizedSearchCV(estimator = classifier_xgb, param_distributions= parameters, n_iter = 5,
                                   scoring = 'roc_auc', n_jobs = -1, cv=5, verbose =3, error_score = 'raise' )
```

```python
random_search.fit(scaled_x_train, y_train)
```

```python
random_search.best_estimator_
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

## Section 4: Final Model (XGBoost Classifier)

```python
classifier = XGBClassifier(colsample_bytree=0.4,
              gamma=0.0, learning_rate=0.05,max_depth=4,
              min_child_weight=5, n_estimators=500)
classifier.fit(scaled_x_train, y_train)
```

```python
y_pred = classifier.predict(scaled_x_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#Confusion matrix
plt.figure(figsize = (5,5))
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True, cmap = 'Set2', fmt= "d",
            linewidth = 3, cbar = False, xticklabels = ['Yes', 'No'],
            yticklabels =['Yes', 'No'], square = True
            )
plt.xlabel("True Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.ylabel("Predicted Labels", fontsize = 12, color = "tab:pink", weight = "bold")
plt.title("CONFUSION MATRIX - XGBOOST", fontsize = 14, color = 'm')
plt.show()
```
